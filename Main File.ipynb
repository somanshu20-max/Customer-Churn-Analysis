```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
import seaborn as sns

df = pd.read_csv('customer churn.csv')
df.head ()
```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
        
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
ax = sns.countplot(x ='Churn', data = df )

ax.bar_label(ax.containers[0])
ply.title("Count of Customer by Churn",fontsize = 15)
ply.show ()

```

```python
ply.figure(figsize = (3,4))
gb= df.groupby("Churn").agg({'Churn':"count"})
ply.pie(gb['Churn'], labels = gb.index,autopct = "%1.2f%%")
ply.title("Percentage of Churn Customer",fontsize = 15)
ply.show()
gb
```

#from the given pie chart we can calculate that 26.54% of our customers have churned out. #not let's explore the reason behind it

```python
ply.figure(figsize =(4,4) )
sns.countplot(x = "gender", data = df, hue = "Churn")
ply.title("Churn by Gender",fontsize = 15)
ply.show()
```

```python
ply.figure(figsize =(4,4) )
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
ply.title("Count of Customer by SeniorCitizen",fontsize = 15)
ply.show()
```

```python
# Step 1: Calculate percentages
data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().fillna(0)
percentages = data.div(data.sum(axis=1), axis=0) * 100

# Step 2: Plot
fig, ax = ply.subplots(figsize=(5, 5))

# Plot each part of the stack
bottom = None
for churn_status in percentages.columns:
    ax.bar(percentages.index, percentages[churn_status],
           bottom=bottom, label=churn_status)
    if bottom is None:
        bottom = percentages[churn_status]
    else:
        bottom += percentages[churn_status]

# Add percentage labels
for i, (index, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for churn_status in percentages.columns:
        value = row[churn_status]
        ax.text(i, cumulative + value / 2, f'{value:.1f}%',
                ha='center', va='center', fontsize=10, color='white')
        cumulative += value

# Aesthetics
ax.set_xticks([0, 1])
ax.set_xticklabels(['Not Senior Citizen', 'Senior Citizen'])
ax.set_ylabel("Percentage")
ax.set_title("Churn by Senior Citizen (Stacked %)", fontsize=14)
ax.legend(title= "Churn", bbox_to_anchor = (1.4,1.4) )

ply.tight_layout()
ply.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Setup grid size (3x3 in this case)
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols  # Round up

# Create figure
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 12))
axes = axes.flatten()  # Flatten in case of multiple rows

# Create a countplot for each column
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue='Churn', ax=axes[i])
    axes[i].set_title(f'{col} vs Churn', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#This visual shows countplots comparing service features with customer churn status across various categories (e.g., PhoneService, InternetService, StreamingTV, etc.). In general, customers without internet-related services or security features tend to churn less, while those with services like Fiber optic or no protection tend to churn more.
